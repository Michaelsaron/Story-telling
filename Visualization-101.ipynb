```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Exercise 1: Monthly Sales Trend

```python
#line chart showing monthly sales trends with green dashed line and square markers
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

plt.plot(months, sales, color='green', linestyle='--', marker='s', markersize=8, label='Sales')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

Exercise 2: Student Scores

```python
#line chart showing exam scores by subject
subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
scores = [85, 78, 92, 88]

plt.plot(subjects, scores, color= 'red', marker= '*', linewidth=1, markerfacecolor= 'green', label= 'Scores')
plt.title('Exam Scores')
plt.xlabel('Subject')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

Exercise 3: Company Revenue

```python
#line chart showing revenue growth over years with purple dotted line and diamond markers
years = [2021, 2022, 2023, 2024]
revenue = [500, 700, 850, 1000]

plt.plot(years, revenue, color= 'purple', marker='D', linestyle=':', label= 'revenue')
plt.title('Revenue Growth Over Years')
plt.xlabel('Years')
plt.ylabel('revenue')
plt.legend()
plt.show()
```

Exercise 4: Fruit Supply (Vertical Bar Chart)

```python
fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]
colors = ['red', 'orange', 'yellow', 'green']

fig, ax = plt.subplots()
bars = ax.bar(fruits, counts, color=colors, label='Fruits')

ax.set_title('Fruit Supply')
ax.set_xlabel('Fruit')
ax.set_ylabel('Quantity')
ax.legend()
plt.show()

```

Class exercise

```python

sales = pd.DataFrame({
    "day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "cups_sold": [120, 135, 110, 145, 160, 210, 190]
})

highest_day = sales.loc[sales["cups_sold"].idxmax()]
print(f"Highest: {highest_day['day']} with {highest_day['cups_sold']} cups")

plt.bar(sales["day"], sales["cups_sold"], color=['blue', 'blue', 'blue', 'blue', 'blue', 'gold', 'orange'])
plt.title('Coffee Sales by Day')
plt.xlabel('Day')
plt.ylabel('Cups Sold')
plt.show()
```

Exercise 5: Stacked Bar Chart (Students by Gender)

```python
departments = ('CS', 'IS', 'SE', 'IT')

students = {
    'Male': np.array([70, 65, 80, 50]),
    'Female': np.array([60, 55, 70, 45])
}

fig, ax = plt.subplots()

bar1 = ax.bar(departments, students['Male'], width=0.7, label='Male')
bar2 = ax.bar(departments, students['Female'], width=0.7, bottom=students['Male'], label='Female')

ax.bar_label(bar1, label_type='center')
ax.bar_label(bar2, label_type='center')

ax.set_title('Students by Department and Gender')
ax.legend()

plt.show()
```

Bonus Challenge: Build a Dashboard

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

#Chart one Monthly Sales Trend 
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

axes[0,0].plot(months, sales, color='green', linestyle='--', marker='s', markersize=8, label='Sales')
axes[0, 0].set_title('Monthly Sales Trend')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Sales')
axes[0, 0].legend()

#Chart two Fruit Supply 
fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]
colors = ['red', 'orange', 'yellow', 'green']

bars = axes[0, 1].bar(fruits, counts, color=colors, label='Fruits')

axes[0,1].set_title('Fruit Supply')
axes[0,1].set_xlabel('Fruit')
axes[0,1].set_ylabel('Quantity')
axes[0,1].bar_label(bars)
axes[0,1].legend()

#Chart three Movie Ratings
movies = ['Movie A', 'Movie B', 'Movie C', 'Movie D']
ratings = [8.5, 7.8, 9.2, 6.5]

bars = axes[1, 0].barh(movies, ratings, color=['gold', 'silver', 'gold', 'brown']) 

axes[1, 0].set_title('Movie Ratings')
axes[1, 0].set_xlabel('Rating')
axes[1, 0].set_ylabel('Movie')
axes[1, 0].bar_label(bars)
axes[1, 0].legend(['Ratings'])

#Chart four Student Distribution by Gender 
departments = ('CS', 'IS', 'SE', 'IT')
male = np.array([70, 65, 80, 50])
female = np.array([60, 55, 70, 45])

bar1 = axes[1, 1].bar(departments, male, width=0.7, label='Male')
bar2 = axes[1, 1].bar(departments, female, width=0.7, bottom=male, label='Female')

axes[1, 1].set_title('Students by Department and Gender')
axes[1, 1].set_xlabel('Department')
axes[1, 1].set_ylabel('Number of Students')
axes[1, 1].bar_label(bar1, label_type='center')
axes[1, 1].bar_label(bar2, label_type='center')
axes[1, 1].legend()

#layout and display
plt.tight_layout()
plt.show()
```
